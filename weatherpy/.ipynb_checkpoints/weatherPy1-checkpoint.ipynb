```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
responses_json = []


# Loop through the list of cities and perform a request for data on each
for city in range(len(cities)):
    print(f"Making request number: {city} for ID: {cities[city]}")
    try:

        responses = requests.get(query_url + str(cities[city])).json()
        name=(responses['name'])
        lat=(responses["coord"]["lat"])
        lon=(responses["coord"]["lon"])
        cloud=(responses["clouds"]["all"])
        temp=(responses["main"]["temp_max"])
        humidity=(responses["main"]["humidity"])
        wind=(responses["wind"]["speed"])
        country=(responses['sys']['country'])
        responses_json.append({"City Name":name, "Latitude":lat,"Longitude":lon, "Cloudiness":cloud,"Temperature":temp,
                    "Humidity":humidity,"Wind Speed":wind,"Country":country})
    except:
        print("An error occurred")
    
    

```

```python
# create a data frame
responses_df = pd.DataFrame(responses_json)
responses_df
```

## Linear Regression

```python
import scipy.stats as st
```

```python
# Create Northern and Southern Hemisphere DataFrames
weather_data
```

```python
NorthHem_df = (weather_data[weather_data["Latitude"] >0]) 
NorthHem_df
```

```python
SouthHem_df = (weather_data[weather_data["Latitude"] <0]) 
SouthHem_df
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(NorthHem_df['Latitude'],NorthHem_df["Temperature"])
regress_values = NorthHem_df['Latitude'] * slope + intercept
plt.scatter(NorthHem_df['Latitude'],NorthHem_df["Temperature"] )
plt.plot(NorthHem_df['Latitude'],regress_values,"r-")
plt.title("Northern Hemisphere City Latitude Versus Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(SouthHem_df['Latitude'],SouthHem_df["Temperature"])
regress_values = SouthHem_df['Latitude'] * slope + intercept
plt.scatter(SouthHem_df['Latitude'],SouthHem_df["Temperature"] )
plt.plot(SouthHem_df['Latitude'],regress_values,"r-")
plt.title("Southern Hemisphere City Latitude Versus Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
```

```python
# As expected, the closer to the equator, the higher the temperature.
```
